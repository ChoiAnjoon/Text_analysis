```python
import pandas as pd 
word_count = pd.read_excel('nsmc_count.xlsx')
```

```python
word_count.head()
```

```python
word_count.sort_values('빈도', ascending=False).head()
```

```python
# 딕셔너리 형태로 변환 
word_dic = word_count.set_index('단어')['빈도'].to_dict()
```

```python
from wordcloud import WordCloud
```

```python
wc = WordCloud(font_path='C:/Windows/Fonts/Malgun.ttf', background_color='white', width=400, height=300)
```

```python
cloud = wc.generate_from_frequencies(word_dic)
```

```python
cloud.to_image()
```

```python
wc = WordCloud(font_path='C:/Windows/Fonts/Malgun.ttf', background_color='white', width=400, height=300)
```

```python
cloud = wc.fit_words(word_dic)
```

```python
cloud.to_image()
```
